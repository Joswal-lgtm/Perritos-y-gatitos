```python
import os
#Importamos el módulo "os" para poder interactuar con nuestro sistema operativo.
os.environ['TF_USE_LEGACY_KERAS'] = '1'
#Usamos este código para que "Tensorflow" acceda a la función "Legacy" del keras.

import numpy as np
#Lo utilizamos para abrir la librería numpy.

from tensorflow import keras
#Carga el modelo keras.
mi_modelo = keras.models.load_model("/content/drive/MyDrive/Colab Notebooks/keras_model.h5", compile=False)
#Carga las etiquetas de cada clase (0-Gatos, 1-Perros).
nombre_clases = open("/content/drive/MyDrive/Colab Notebooks/labels.txt", "r").readlines()
#Usamos print para cerciorarnos de que el código funciona bien al mostrar las etiquetas.
print(nombre_clases)

from PIL import Image, ImageOps
imagen= Image.open("/content/drive/MyDrive/Colab Notebooks/Cats_Dogs_test folder/gato_dibujo_1.jpeg").convert("RGB")
size = (224, 224)
imagen = ImageOps.fit(imagen, size, Image.Resampling.LANCZOS)
imagen_array = np.asarray(imagen)
normalizada_imagen_array = (imagen_array.astype(np.float32) / 127.5) - 1
print(imagen_array.shape)

lote_imagenes = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
lote_imagenes[0] = normalizada_imagen_array
resultados= mi_modelo.predict(lote_imagenes)
print(resultados)

indice = np.argmax(resultados[0])
etiqueta = nombre_clases[indice]
probabilidad = resultados[0][indice]
print("La imagen es de clase: ", etiqueta)
print("Con una probabilidad de: ", probabilidad)

import os
lista_archivos = os.listdir("/content/drive/MyDrive/Colab Notebooks/Cats_Dogs_test folder")
total_predicciones = 0
aciertos = 0
media_probabilidad_aciertos = 0.0
predicciones_incorrectas = []
print(aciertos)

from PIL import Image, ImageOps
import numpy as np

def predecir_imagen(ruta_imagen):
    imagen = Image.open(ruta_imagen).convert("RGB")
    size = (224, 224)
    imagen = ImageOps.fit(imagen, size, Image.Resampling.LANCZOS)
    imagen_array = np.asarray(imagen)
    normalizada_imagen_array = (imagen_array.astype(np.float32) / 127.5) - 1

    lote_imagenes = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
    lote_imagenes[0] = normalizada_imagen_array
    resultados = mi_modelo.predict(lote_imagenes, verbose=0)

    indice = np.argmax(resultados[0])
    etiqueta_predicha_raw = nombre_clases[indice].strip()

    if ' ' in etiqueta_predicha_raw:
        etiqueta_predicha = etiqueta_predicha_raw.split(' ')[1].lower()
    else:
        etiqueta_predicha = etiqueta_predicha_raw.lower()

    probabilidad = resultados[0][indice]

    return etiqueta_predicha, probabilidad

cat_label_str = "cat"
dog_label_str = "dog"
image_folder_path = "/content/drive/MyDrive/Colab Notebooks/Cats_Dogs_test folder"

for nombre_archivo in lista_archivos:
    etiqueta_esperada = None
    if cat_label_str in nombre_archivo.lower():
        etiqueta_esperada = "gatos"
    elif dog_label_str in nombre_archivo.lower():
        etiqueta_esperada = "perros"
    else:
        continue

    total_predicciones += 1

    ruta_imagen = os.path.join(image_folder_path, nombre_archivo)
    etiqueta_predicha, probabilidad = predecir_imagen(ruta_imagen)

    if etiqueta_predicha == etiqueta_esperada:
        aciertos += 1
        media_probabilidad_aciertos += probabilidad
    else:
        info_error= {"archivo": nombre_archivo, "prediccion": etiqueta_predicha,
                       "probabilidad": probabilidad, "esperado": etiqueta_esperada}
        predicciones_incorrectas.append(info_error)

if aciertos >0:
  precision = (aciertos/total_predicciones) * 100
  print(f"Precisión: {precision:2f}%")
else:
  print("No hay imágenes para procesar")

precision = aciertos / total_predicciones if total_predicciones > 0 else 0
media_probabilidad_aciertos /= aciertos if aciertos > 0 else 1
informe = f"""
Informe de Evaluación del Modelo
Total de predicciones: {total_predicciones}
Aciertos: {aciertos}
Precisión: {precision:.2%}
Probabilidad media de aciertos: {media_probabilidad_aciertos:.2%}
Predicciones incorrectas:
"""

for error in predicciones_incorrectas:
    informe += f"Archivo: {error['archivo']}, Predicción: {error['prediccion']}, Probabilidad: {error['probabilidad']:.2%}, Esperado: {error['esperado']}\n"

precision = aciertos / total_predicciones if total_predicciones > 0 else 0
media_probabilidad_aciertos /= aciertos if aciertos > 0 else 1

informe = f"""
Informe de Evaluación del Modelo

Total de predicciones: {total_predicciones}
Aciertos: {aciertos}
Precisión: {precision:.2%}
Probabilidad media de aciertos: {media_probabilidad_aciertos:.2%}

Predicciones incorrectas:
"""

for error in predicciones_incorrectas:
    informe += f"Archivo: {error['archivo']}, Predicción: {error['prediccion']}, Probabilidad: {error['probabilidad']:.2%}\n"

print(informe)
```
